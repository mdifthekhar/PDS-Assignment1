```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as pl
from ISLP.models import (ModelSpec as MS,summarize,poly)
fratlity_data=pd.read_csv('C:/Users/Moham/OneDrive/Documents/PDS_ASSIGNMENT1/1/clean_data/clean_data.csv')
```

```python
fratlitydf=pd.DataFrame(fratlity_data)
fratlitydf
```

```python
prefratiliy=fratlitydf.drop(columns=['Unnamed: 0'])
```

```python
prefratiliy
```

```python
predictor=MS(['Age']).fit_transform(prefratiliy)
response=prefratiliy['Frailty']
model=sm.OLS(response,predictor).fit()
model.summary()
```

```python
x=prefratiliy['Age']
y=prefratiliy['Frailty']
pl.xlabel('Age')
pl.ylabel('Frailty')
pl.legend()
pl.bar(x,y)
pl.savefig('C:/Users/Moham/OneDrive/Documents/PDS_ASSIGNMENT1/1/analysis/test_result.png')
```
